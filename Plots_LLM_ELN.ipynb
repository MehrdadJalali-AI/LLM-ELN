```python
import matplotlib.pyplot as plt
import numpy as np

# Data
questions = ['Q1', 'Q2', 'Q3', 'Q4']
labels = ['LLM-ELN vs. Expert', 'LLM-ELN vs. GPT', 'Expert vs. GPT']

# Similarity Scores for Each Question
cosine_similarities = [
    [0.1575, 0.4597, 0.1190],  # Q1
    [0.1575, 0.1898, 0.0660],  # Q2
    [0.1575, 0.4597, 0.1190],  # Q3
    [0.2953, 0.4011, 0.2952]   # Q4
]

semantic_similarities = [
    [0.6978, 0.8583, 0.5611],  # Q1
    [0.7768, 0.8573, 0.8121],  # Q2
    [0.6978, 0.8583, 0.5611],  # Q3
    [0.8226, 0.9384, 0.9015]   # Q4
]

# Function to plot spider chart
def plot_spider_chart(data, question, metric, labels):
    categories = labels
    N = len(categories)
    angles = [n / float(N) * 2 * np.pi for n in range(N)]
    angles += angles[:1]

    ax = plt.subplot(111, polar=True)
    plt.title(f'{question} - {metric} Similarity Distribution', loc='left')
    ax.set_theta_offset(np.pi / 2)
    ax.set_theta_direction(-1)

    plt.xticks(angles[:-1], categories)

    ax.set_rlabel_position(0)
    plt.yticks([0.25, 0.5, 0.75], ["0.25", "0.50", "0.75"], color="grey", size=7)
    plt.ylim(0, 1)

    values = data[0]
    values += values[:1]
    ax.plot(angles, values, linewidth=1, linestyle='solid', label=question)
    ax.fill(angles, values, 'b', alpha=0.1)

    plt.show()

# Plot Cosine Similarities
for i, scores in enumerate(cosine_similarities):
    plot_spider_chart([scores], questions[i], "Cosine", labels)

# Plot Semantic Similarities
for i, scores in enumerate(semantic_similarities):
    plot_spider_chart([scores], questions[i], "Semantic", labels)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
questions = ['LLM-ELN vs. Expert', 'LLM-ELN vs. GPT', 'Expert vs. GPT']
labels = questions  # These will be the axes in the radar chart

# Similarity Scores
cosine_similarities = [0.1575, 0.1575, 0.2953]
semantic_similarities = [0.6978, 0.7768, 0.8226]

# Combined data
data = [cosine_similarities, semantic_similarities]
metrics = ['Cosine Similarity', 'Semantic Similarity']

# Function to plot radar chart
def plot_radar_chart(data, labels, metrics):
    N = len(labels)

    # Setup angles for radar plot
    angles = np.linspace(0, 2 * np.pi, N, endpoint=False).tolist()
    angles += angles[:1]

    # Setup plot
    fig, ax = plt.subplots(figsize=(8, 8), subplot_kw=dict(polar=True))
    ax.set_theta_offset(np.pi / 2)
    ax.set_theta_direction(-1)

    plt.xticks(angles[:-1], labels)

    colors = ['#1f77b4', '#ff7f0e']

    # Plot each metric
    for i, scores in enumerate(data):
        values = scores + scores[:1]
        ax.plot(angles, values, linewidth=2, linestyle='solid', label=metrics[i], color=colors[i % len(colors)])
        ax.fill(angles, values, color=colors[i % len(colors)], alpha=0.2)

    # Add title and legend
    plt.title('Cosine & Semantic Similarity Distribution', fontsize=15, fontweight='bold')
    ax.legend(loc='upper right', bbox_to_anchor=(1.1, 0.1), fontsize=12)
    plt.show()

# Plot the combined radar chart
plot_radar_chart(data, labels, metrics)


```
